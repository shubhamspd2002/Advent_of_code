```python
with open('input.txt', 'r') as f:
    shelf = f.read().splitlines()

    # Following does padding of . around the text matrix to make a new text matrix. This is a list within a list
    new_shelf_list = []
    new_shelf_list.append(['.' for i in range(len(shelf[0]))])
    for s in shelf:
        new_shelf_list.append(['.'] + [letter for letter in s] + ['.'])
    new_shelf_list.append(['.' for i in range(len(shelf[0]))])
    # print(new_shelf_list)
    count_of_X = 0
    old_count_of_X = 0
    while True:
        old_count_of_X = count_of_X
        for row_idx, row in enumerate(new_shelf_list):
            for roll_idx, roll in enumerate(row):
                if roll == '.':
                    continue
                else:
                    neighbours_of_roll = list()
                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx-1][roll_idx])
                    except IndexError as e:
                        _ = e

                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx-1][roll_idx-1])
                    except IndexError as e:
                        _ = e

                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx-1][roll_idx+1])
                    except IndexError as e:
                        _ = e
                        

                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx+1][roll_idx])
                    except IndexError as e:
                        _ = e
                    
                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx+1][roll_idx-1])
                    except IndexError as e:
                        _ = e
                    
                    try:
                        neighbours_of_roll.append(new_shelf_list[row_idx+1][roll_idx+1])
                    except IndexError as e:
                        _ = e

                    if neighbours_of_roll.count('@') > 3: # early skipping
                        continue

                    try:
                        neighbours_of_roll.append(row[roll_idx-1])
                    except IndexError as e:
                        _ = e

                    try: 
                        neighbours_of_roll.append(row[roll_idx+1])
                    except IndexError as e:
                        _ = e

                    if neighbours_of_roll.count('@') > 3: 
                        continue
                    else:
                        new_shelf_list[row_idx][roll_idx] = '.'
                        count_of_X += 1
        if old_count_of_X == count_of_X:
            break

print(count_of_X)

        

    
```
