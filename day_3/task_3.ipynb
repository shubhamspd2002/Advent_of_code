```python
with open('input.txt', 'r') as f:
    f = f.read().splitlines()
    voltage = 0
    initial_idx = 1
    for i in f:
        # print(i)
        max_number_in_order = 0
        for j in i:
            for k in i[initial_idx:]:
                number = int(j+k)
                # print(number)
                if number >= max_number_in_order:
                    max_number_in_order = number
            initial_idx += 1
        voltage += max_number_in_order
        initial_idx = 1
            

print(voltage)
```

```python
with open('ip.txt', 'r') as f:
    f = f.read().splitlines()
    voltage = 0
    for S in f:
        stack = []
        k = 12
        for idx, s in enumerate(S):
            d = int(s)
            # What the following while loop means:
            # If stack has elements (to pop) and the new number is greater than the last number in stack and incase we pop from the stack (len(stack) - 1) + the remaining numbers  to be iterated upon  (len(S) - idx) >= k, 
            # then we have more numbers to iterate upon to find a better one and as the current one is greater, just pop from the stack. Following is one example:
            # d = 4, stack = [9, 2, 3] and s = 92345565565343223 then this while loop removes the 2 and 3 as they are smaller than 4 and we have more numbers to explore. 
            # We exit the while loop cause 4<9 and we have stack = [9] and the 4 will get appended later in the if block
            while stack and stack[-1] < d and len(stack) - 1 + (len(S) - idx) >= k: 
                _ = stack.pop()
            # this if block executes if we have more space to put the digits regardless of whether they are optimal choice or not. The ellimination is done by the while loop.
            if len(stack) < k: 
                stack.append(d)
            # If the length is equal to k then continue ahead becuase there might be numbers that are greater than stack[-1]
            if len(stack) == k: 
                continue
        # print(stack)
        max_voltage_of_a_bank = int(''.join([str(a) for a in stack]))
        # print(max_voltage_of_a_bank)
        voltage += max_voltage_of_a_bank




print(voltage)
```
